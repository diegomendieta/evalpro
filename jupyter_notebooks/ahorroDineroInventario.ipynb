```python
valor_por_sku = pd.read_csv('../data/valorSku.csv', error_bad_lines=False, encoding='UTF-8')
```

```python
conversion = pd.read_csv('../data/conversiones_unidades.csv', encoding = "ISO-8859-1")[['ID_SKU_VENTA', 'CAJAS_POR_PALLET']]
sku_pallets = {}
for index, row in conversion.iterrows():
    sku_pallets[row['ID_SKU_VENTA']] = row['CAJAS_POR_PALLET']
valor_por_sku['Precio Pallet'] = valor_por_sku.apply(lambda x: x['Precio Caja'] * sku_pallets[x['SKU']], axis=1)
```
