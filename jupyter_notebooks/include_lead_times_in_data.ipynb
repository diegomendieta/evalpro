```python
lead_times_dict = {}

for index, row in lead_times.iterrows():
    
    lead_times_dict[(row['Origen'], row['Destino'])] = row['LDT [Días]']
```

```python
mean_lead_times = lead_times['LDT [Días]'].mean()
```

```python
mean_lead_times
```

```python
def assign_lead_time(x, lead_times_dict, mean_lead_times):
    if (x['HUB'], x['DESCR_CENDIST']) in lead_times_dict:
        return  lead_times_dict[(x['HUB'], x['DESCR_CENDIST'])]
    elif  x['HUB'] == x['DESCR_CENDIST']:
        return 0
    else:
        return mean_lead_times
```

```python
all_data['lead_time'] = all_data.apply(lambda x: assign_lead_time(x, lead_times_dict, mean_lead_times), axis=1)
```
