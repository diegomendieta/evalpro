```python
import pandas as pd

equivalencias = pd.read_csv('data/equivalencia_cajas.csv', encoding = "ISO-8859-1")
stock_saves = pd.read_csv('data/delta_s_por_sku.csv')
```

```python
data = stock_saves.merge(equivalencias)
```

```python
data['INVENTARIO_AHORRADO_PROMEDIO'] = data.apply(lambda x: (x['SUM(PREVENTA)'] * (x['CAPAS_X_PALLET'] - 1)) / 2 if x['CAPAS_X_PALLET'] > 1 else x['SUM(PREVENTA)'], axis=1)
```

```python
data
```

```python
data = data.drop(columns=['DESCR_SKU_LONG', 'CAJAS_POR_PALLET', 'CAJAS_POR_CAPA', 'SUM(PREVENTA)', 'CAPAS_X_PALLET'], axis=1)
```

```python
with open('data/delta_inventario_promedio_por_sku.csv', 'w') as file:
    file.write('ID_SKU_VENTA' + ',' + 'INVENTARIO_PROMEDIO_AHORRADO' + '\n')
    
    for index, row in data.iterrows():
        file.write(str(row['ID_SKU_VENTA']) + ',' + str(row['INVENTARIO_AHORRADO_PROMEDIO']) + '\n')
```
