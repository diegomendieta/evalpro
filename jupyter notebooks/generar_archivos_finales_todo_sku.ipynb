```python
import pandas as pd
from datetime import datetime
from datetime import date
import json
import math
import xlrd
import numpy as np

datos_no_agrupados = pd.read_csv('../data/stats_no_agrupados_limpios.csv')
datos_agrupados = pd.read_csv('../stats_sku_entero/stats_agrupados_limpios_100%_90%conf.csv')
datos_no_agrupados = datos_no_agrupados.drop(columns='Unnamed: 0')
datos_agrupados = datos_agrupados.drop(columns='Unnamed: 0')
```

```python
datos_finales = datos_no_agrupados.merge(datos_agrupados, on=['ID_SKU_VENTA', 'DESCR_CENDIST'], how='outer')
datos_finales = datos_finales.fillna(0)
```

```python
datos_finales
```

```python
datos_finales_utiles = datos_finales[['ID_SKU_VENTA', 'DESCR_CENDIST', 'STOCK_TEORICO_x', 'STOCK_TEORICO_y']]
datos_finales_utiles['DELTA'] = datos_finales_utiles['STOCK_TEORICO_y'] - datos_finales_utiles['STOCK_TEORICO_x']
datos_finales_utiles = datos_finales_utiles.drop(columns=['STOCK_TEORICO_x', 'STOCK_TEORICO_y'])
```

```python
datos_finales_utiles.to_csv('../deltas/delta_inventario_sku_cd_100%_90%conf.csv')
```
