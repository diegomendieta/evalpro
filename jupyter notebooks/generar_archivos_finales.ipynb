```python
import pandas as pd
from datetime import datetime
from datetime import date
import json
import math
import xlrd
import numpy as np

datos_no_agrupados = pd.read_csv('../data/stats_no_agrupados_limpios.csv')
datos_agrupados = pd.read_csv('../data/stats_agrupados_limpios.csv')
datos_no_agrupados = datos_no_agrupados.drop(columns='Unnamed: 0')
datos_agrupados = datos_agrupados.drop(columns='Unnamed: 0')
```

```python
datos_no_agrupados 
```

```python
datos_agrupados
```

```python
datos_finales = datos_no_agrupados.merge(datos_agrupados, on=['ID_SKU_VENTA', 'DESCR_CENDIST'], how='outer')
```

```python
datos_finales = datos_finales.fillna(0)
```

```python
datos_finales
```

```python
datos_finales_agrupados = datos_finales[['ID_SKU_VENTA', 'DESCR_CENDIST', 'MEDIA_x', 'SS_y']]
```

```python
datos_finales_agrupados = datos_finales_agrupados.rename(columns={'MEDIA_x': 'MEDIA', 'SS_y': 'SS'})
```

```python
datos_finales_agrupados
```

```python
datos_finales_agrupados['INVENTARIO_TOTAL'] = datos_finales_agrupados.apply(lambda x: x.MEDIA + x.SS, axis=1)
```

```python
datos_finales_agrupados.to_csv('../data_variabilidad_agrupada/datos_agrupados.csv')
```

```python
datos_no_agrupados = datos_no_agrupados[['ID_SKU_VENTA', 'DESCR_CENDIST', 'MEDIA', 'SS', 'STOCK_TEORICO']].rename(columns={'STOCK_TEORICO': 'INVENTARIO_TOTAL'})
```

```python
datos_no_agrupados.to_csv('../data_variabilidad_agrupada/datos_no_agrupados.csv')
```

```python
data_diferencias = datos_finales_agrupados.merge(datos_no_agrupados, on=['ID_SKU_VENTA', 'DESCR_CENDIST'], how='outer')
```

```python
data_diferencias = data_diferencias.fillna(0)
```

```python
data_diferencias['DELTA_INVENTARIO'] = data_diferencias.apply(lambda x: x['INVENTARIO_TOTAL_x'] - x['INVENTARIO_TOTAL_y'], axis=1)
```

```python
data_diferencias = data_diferencias[['ID_SKU_VENTA', 'DESCR_CENDIST', 'DELTA_INVENTARIO']]
```

```python
data_diferencias.to_csv('../data_variabilidad_agrupada/delta_inventario_sku_cd.csv')
```

```python
valor_por_sku = pd.read_csv('../data/Valor por SKU.csv', encoding = "ISO-8859-1")
```

```python
valor_por_sku = valor_por_sku[['SKU', 'Precio Caja']]
```

```python
valor_por_sku
```

```python
conversion = pd.read_csv('../data/conversiones_unidades.csv', encoding = "ISO-8859-1")[['ID_SKU_VENTA', 'CAJAS_POR_PALLET']]
sku_pallets = {}
for index, row in conversion.iterrows():
    sku_pallets[row['ID_SKU_VENTA']] = row['CAJAS_POR_PALLET']
```

```python
valor_por_sku['Precio Pallet'] = valor_por_sku.apply(lambda x: x['Precio Caja'] * sku_pallets[x['SKU']], axis=1)
    
```

```python
valor_por_sku
```

```python
valor_por_sku = valor_por_sku.rename(columns={'SKU': 'ID_SKU_VENTA'})
data_diferencias_precios = data_diferencias.merge(valor_por_sku, on='ID_SKU_VENTA')
```
